# Import libraries

```python
# import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
# import scipy as sp

plt.rcParams["font.family"] = "Helvetica"
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

# For one-column paper

```python
X = np.linspace(0, 10*np.pi, 1000)
Y = np.cos(X)
X[::100], Y[::100]
```

```python
fig_size = tuple(np.array([15, 5])/2.54)

fig, ax = plt.subplots(1, 1, figsize=fig_size)
ax.plot(X, Y, color='b')
ax.set_xlim([np.min(X), np.max(X)])
ax.set_ylim([np.min(Y), np.max(Y)])
ax.set_xlabel(f'Frequency ($\omega a/2\pi c$)')
ax.set_ylabel(f'Transmission (T)')

fig.tight_layout()
fig.savefig('./figs/one_column_py.pdf')
```
